```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
product_name = []
price = []
description = []
rating = []


for i in range(1,8):
    url = f"https://www.flipkart.com/search?q=phone&as=on&as-show=on&otracker=AS_Query_PredictiveAutoSuggest_7_0_na_na_na&otracker1=AS_Query_PredictiveAutoSuggest_7_0_na_na_na&as-pos=7&as-type=PREDICTIVE&suggestionId=phone&requestId=778f5b7d-9c17-4a78-ba10-02f6efc27ab5&page={i}"
    r = requests.get(url)

    soup = BeautifulSoup(r.content,"html.parser")
    box = soup.find("div", class_="_1YokD2 _3Mn1Gg")


    names = box.find_all("div", class_= "_4rR01T")
    for i in names:
        d=i.text
        product_name.append(d)
    

    prices= box.find_all("div", class_="_30jeq3 _1_WHN1")
    for i in prices:
        d=i.text
        price.append(d)
  

    descriptions = box.find_all("ul", class_="_1xgFaf")
    for i in descriptions:
        d=i.text
        description.append(d)
    

    ratings = box.find_all("div", class_="_3LWZlK")
    for i in ratings:
        d=i.text
        rating.append(d)
```

```python
df = pd.DataFrame({"Product Name":product_name, "Price":price, "Description":description, "Rating":rating})
df
```
